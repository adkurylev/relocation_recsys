Parser for tabular data from wiki

```python
import numpy as np
import pandas as pd
import requests 
from bs4 import BeautifulSoup
```

```python
def temp_2_float(text):
    try:
        num = text.split('(')[0]
        if '−' in num:
            num = num.replace('−', '-')
        num = float(num)
        return num
    except:
        #print(text)
        return np.nan
```

```python
# get the response in the form of html
wikiurl="https://en.wikipedia.org/wiki/List_of_cities_by_average_temperature"
table_class="wikitable plainrowheaders sortable jquery-tablesorter"
response=requests.get(wikiurl)
print(response.status_code)
```

```python
# parse data from the html into a beautifulsoup object
soup = BeautifulSoup(response.text, 'html.parser')
wikitables = soup.findAll('table', {'class':"wikitable"})
tables = []

for table in wikitables:
    # convert list to dataframe
    df = pd.read_html(str(table))[0]
    cols = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec', 'Year']
    df[cols] = df[cols].applymap(temp_2_float)
    df = df.drop(df.iloc[:, len(df.columns)-1:], axis = 1)
    tables.append(df)

wiki_df = pd.concat(tables)

wiki_df
```
